```python
1. Load the Data
> •  Load the dataset using Pandas
> • Display the first 10 rows
> • Display the last 10 rows
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('train.csv')

# Display the first 10 rows
print("First 10 Rows:")
display(df.head(10))

# Display the last 10 rows
print("\nLast 10 Rows:")
display(df.tail(10))
```

```python
2. Data Exploration
> • Number of rows and columns
> • Data types
> • Check for missing values
```

```python
# Number of rows and columns
print(f"Shape of the dataset: {df.shape}")

# Data types
print("\nData Types:")
print(df.dtypes)

# Check for missing values
print("\nMissing Values per Column:")
print(df.isnull().sum())
```

```python
2.1. Data Cleaning (if needed)
> • Handle missing values
> • Fix or remove incorrect data
```

```python
# Handling missing values (Filling 'Arrival Delay' with mean)
if 'Arrival Delay in Minutes' in df.columns:
    df['Arrival Delay in Minutes'] = df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].mean())

# Remove unnecessary columns (Fixing data)
df.drop(['Unnamed: 0', 'id'], axis=1, inplace=True, errors='ignore')

# Final check
print("Remaining missing values:", df.isnull().sum().sum())
```

```python
3. Statistical Analysis
> Using Pandas and NumPy, calculate:
> • Mean, Median, Standard deviation, Minimum and maximum values
```

```python
# Select only numeric columns for analysis
numeric_df = df.select_dtypes(include=[np.number])

# Calculate Statistics
stats_summary = pd.DataFrame({
    'Mean': numeric_df.mean(),
    'Median': numeric_df.median(),
    'Standard Deviation': numeric_df.std(),
    'Minimum': numeric_df.min(),
    'Maximum': numeric_df.max()
})

display(stats_summary)
```

```python
4. Data Visualization (Matplotlib)
> Create at least 3 plots: Bar chart, Line chart, Histogram
```

```python
# 1. Bar Chart (Satisfaction Count)
plt.figure(figsize=(8, 5))
df['satisfaction'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Bar Chart: Satisfaction Levels')
plt.xlabel('Satisfaction Status')
plt.ylabel('Count')
plt.show()

# 2. Histogram (Age Distribution)
plt.figure(figsize=(8, 5))
plt.hist(df['Age'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Histogram: Distribution of Passenger Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 3. Line Chart (Flight Distance Sample)
plt.figure(figsize=(10, 5))
plt.plot(df['Flight Distance'].iloc[:50], marker='o', linestyle='-', color='orange')
plt.title('Line Chart: Flight Distance (First 50 Samples)')
plt.xlabel('Index')
plt.ylabel('Distance')
plt.show()
```
